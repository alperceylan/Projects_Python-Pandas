### **Measuring Rating Disagreement**
+ Suppose you wanted to find the movies that are most divisive between male and female viewers. 
<br>One way is to add a column to **'mean_ratings'** containing the difference in means, then sort by that **:** 
+ Erkek ve kadın izleyiciler arasında **,** oy farkının en fazla olduğu filmleri bulmak istediğinizi varsayalım.
<br> Bunun bir yolu **,** Ortalamalardaki farkı içeren **'mean_ratings'** Tablosuna bir **"difference" (fark)** Sütunu eklemek ve ardından buna göre sıralamaktır. **(** Eksili değerler de çıkabileceği için en son da **Mutlak Değer**'lerini alalım... **) :**

**+--- NOTE ---> Mutlak Değer** almak için de **,** "**``abs( )``**" fonksiyonunu kullanırız **!**

```python
mean_ratings_active_titles.head()
```

```python
#   Add diff column to "mean_ratings_active_titles" :
#   "ortalama_ratings_active_titles"'a "different" sütunu ekle :

#   Take the absolute value of the "different" column :
#   Yarattığımız "different" sütununun , mutlak değerini al :

mean_ratings_active_titles["different"] = abs(mean_ratings_active_titles["F"] - mean_ratings_active_titles["M"]) ;
mean_ratings_active_titles
```

```python
#   OR

mean_ratings_active_titles["different"] = abs(mean_ratings_active_titles.F - mean_ratings_active_titles.M) ;
mean_ratings_active_titles
```

```python
#   Sort values by "different" column:
#   Değerleri , "different" sütununa göre sırala :
#   Bu bize , kadınlar ve erkekler arasındaki beğeni farkına göre bir sıralama yapacak :

mean_ratings_active_titles.different.sort_values(ascending=False)
```

```python
#   OR

mean_ratings_active_titles.sort_values(by="different" , ascending=False)
```

**++++ NOTE ++++**
<br> **"``.nlargest( )``" ::** Bu fonksiyon ile **;** bütün tabloyu sıralamak yerine **,** mesela **"** kadınlar ve erkekler arasındaki beğeni farkının EN FAZLA olduğu sadece **6** film **"** gibi bir bilgi isteniyorsa **,** bu fonksiyon daha pratik bir kullanım sağlıyor **:**

#### Bu işlemi **;** başka bir yol olan **,** herhangi bir **"Regular Expression"** metodu ile gerçekleştirelim **:**
<br> (**"``.str``"** fonksiyonu **;** tamamen kendine özgü metodları olduğu gibi **Regular Expression** metodları da içerir)
<br> Mesela bunlardan biri de **, "``.str.findall( )``"** metodu **:**
<br> **[ ``.findall( )``** metodu da **,** bir **``pattern``** belirlenip **,** içerisine yazılarak kullanılıyor... **]**

**+--- NOTE --->** Normalde **parantez** karakterinin **,** STRING ifadelerde PATTERN hazırlarken özel bir anlamı (kullanımı) var... Ama biz direkt **parantez** karakterinin kendisini kullanmak istiyorsak **,** kural içerisinde önüne **" ``\`` "** karakteri koyarak o şekilde yazmalıyız... Yani **, "``\ (``"** şeklinde yazmalıyız...
<br> **+--- NOTE --->** Mesela **,** Parantez sonrasında rakam da olsun istiyorsak... Bunun için de **, "``\d``"** karakterini kullanıyoruz... Rakam sayısının da TAMI TAMINA **4** tane olmasını istiyorsak **"``\d``"** sonrasında **, " ``{4}`` "** yazmamız gerekiyor... Yani sonuçta TAM OLARAK **4** tane rakam olsun istiyorsak **, "``\d{4}``"** şeklinde yazmamız gerekiyor...

```python
data_reduced["year"] = data_reduced.title.str.findall("\(\d{4}\)")
```

```python
data_reduced.head()
```

**++++ NOTE ++++**
<br> **Peki , PATTERN'de PARANTEZ ifadesinin GÖREVİ nedir ?**
<br> Parantez ifadesi **;** MESELA **,** AD **=** Alper **-** SOYAD **=** Ceylan ifadesi gibi bir **eşleştirme** yapmak istiyoruz... Bunun için **, " AD=\w+ "** şeklinde bir PATTERN oluşturmamız gerekiyor... Burada **" AD="** kısmının görünmeyip **,** sadece ismin kendisinin görünmesini istiyorsak **,** bu kuralı **" ( ) "** içerisinde yazmamız gerekiyor... Yani **,** SADECE görünmesini istediğimiz yeri **,** PATTERN'de PARANTEZ içerisine almamız gerekiyor. Yani **, "``AD=(\w+)``"** şeklinde yazmamız gerekiyor... **:**

```python
data_reduced["year"] = data_reduced.title.str.findall("\((\d{4})\)");
```

```python
data_reduced.head()
```

Yukarıda gördüğümüz gibi **,** artık Yıl değeri parantez içerisinde değil... Parantezleri PATTERN'de belirlediğimiz üzere dışarıda bıraktık... Sadece Köşeli Parantez kaldı...

**+--- NOTE --->** OUT olarak **"year"** Kolonunda verilerin Köşeli Parantez içerisinde yazmasının sebebi de **,** "**.findall( )**" metodunun OUT olarak LİSTE döndürmesidir...

+ KÖŞELİ PARANTEZ'den de kurtulmak istiyorsak **,** yani LİSTE içerisine girmeden bu işlemi gerçekleştirmesini istiyorsak **,** yine STRING fonksiyonunun metodlarından birini kullanabiliriz **:**
<br> **( ``.str.extract( )``** metodu **)**

```python
data_reduced["year"] = data_reduced.title.str.extract("\((\d{4})\)")
```
