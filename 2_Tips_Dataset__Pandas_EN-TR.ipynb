##### Yani yaratacağımız fonksiyon **,** Yukarıdaki şekilde Gruplayacak **,** Bahşiş Yüzdesine göre Sıralayacak ve **en fazla bahşiş veren '5' veriyi getirecek :**

##### AŞAĞIDAKİ YARATTIĞIMIZ FONKSİYONUN AÇIKLAMASI **:**

**def** _fonksiyon ismi_ **(** Tablo ismi **, n =** kaç tane veri istiyoruz **, column=** üzerinde çalışmak istediğimiz kolon ismi **) :**
<br> **return** Tablo **.** _küçükten-büyüğe_sıralama_fonksiyonu_ **(** neye_göre_sıralama **=** column **) [** son **n** sayıda veri **]**

```python
#   This function sorts a DataFrame and return the top "n" elements:
#   Yarattığımız bu fonksiyon , bir DataFrameyi sıralar ve ilk "n" öğeyi döndürür :

def top(df , n=5 , column='tip_pct'):
    return df.sort_values(by=column)[-n:]
```

```python
#   OR
#   +--- NOTE ---> Aşağıdaki fonksiyon yazma şekli ve yukarıdaki fonksiyon yazma şekli aynı sonucu verir!

def top(df , n=5 , column='tip_pct'):
    return df.sort_values(by=column , ascending=False)[:n]
```

+ Herhangi bir kolon ismi yazmadan **,** DataFramedeki Tüm Kolonlara bakarak **en çok bahşiş verenlerin yüzdesine göre ilk '4' Satırı sıralayalım :**

```python
top(tips , n=4)
```

+ Şimdi de **;** Sigara İçen ve İçmeyenlere göre **, en çok bahşiş verenlerin yüzdesine göre sıralayalım :**

```python
tips.groupby('smoker').apply(top)
```

**++++ NOTE ++++**
<br> **.groupby( ).apply( )** kullandığımız zaman **groupby ,** tabloları Bütün Halde input olarak gönderiyor. Yani **DataFrame** olarak gönderiyor... ANCAK **DataFrame.apply( )** kullandığımız zaman **, SERIES** halde input olarak gönderiyor... **:**

```python
tips.groupby('smoker').apply(lambda x : type(x))
```

```python
tips.apply(lambda x: str(type(x)))
```

Yukarıda gördüğümüz gibi... **'groupby'** kullanırsak inputlar DataFrame olarak **,** direkt DataFrame'yi kullanırsak da inputlar SERIES olarak gönderiliyor...

**+--- NOTE --->** The **'top'** function is called on each row group from the DataFrame **,** and then the results are glued together using **pandas.concat ,** labeling the pieces with the group names.
<br> **+--- NOTE --->** DataFrame'den her Satır Grubunda **,** daha önce yarattığımız **'top'** fonksiyonu çağırılır ve ardından sonuçlar **'pandas.concat'** kullanılarak birbirine yapıştırılır ve parçalar Grup adlarıyla etiketlenir...

**+--- NOTE --->** If you pass a function to apply that takes other arguments or keywords, you can pass these after the function...
<br> **+--- NOTE --->** Başka bağımsız değişkenler veya anahtar sözcükler alan bir fonksiyonu uygulamak için iletirseniz **,** bunları fonksiyondan sonra iletebilirsiniz...

+ Mesela başka Kolonlara bakmak istersek **,** örneğin **sigara içimine göre en fazla hesap ödeyenlere bakmak istersek :**

```python
tips.groupby('smoker').apply(top(column='total_bill'))
```

**'.apply( )'** fonksiyonunda yukarıdaki gibi bir yazım yapamıyoruz... Fonksiyon ismi gibi **,** diğer argümanları da input olarak yan yana yazmamız gerekiyor. (Yani **,** virgüllerle ayırarak yazmamız gerekiyor.) **:**

```python
tips.groupby('smoker').apply(top , column = 'total_bill')
```

```python
tips.groupby(['smoker' , 'day']).apply(top , n=2 , column='total_bill')
```

# Pivot Tables 

A pivot table is a data summarization tool frequently found in spreadsheet programs and other data analysis software. It aggregates a table of data by one or more keys, arranging the data in a rectangle with some of the group keys along the rows and some along the columns. Pivot tables in Python with pandas are made possible through the groupby facility described in this chapter combined with reshape operations utilizing hierarchical indexing.

**Groupby** mantığı ile aynıdır ancak **Pivot Table**'in içeriği daha kapsamlıdır. **Groupby** ile birkaç fonksiyon ile yapabileceğimiz işi **, Pivot Table** ile tek fonksiyonda yapabiliyoruz...
<br> **+--- NOTE ---> Pivot Table** default olarak verilerin ORTALAMASINI ALIR! (SHIFT+TAB yaparak baktığımızda da **aggfunc='mean'** ifadesini görebiliriz..)

```python
 tips.pivot_table(index=['day' , 'smoker'])
```

+ Aşağıda **, '``.pivot_table( )``'** fonksiyonunun içerisine İLK koyduğumuz elemanı **values** olarak yerleştiriyor. Yani **ilk** koyduğumuz eleman ile **,** sadece **üzerinde işlem yapmak istediğimiz Kolonları** seçiyoruz... Sonrasında yazdıklarımız da **;** hangi başlıkları Kolon olarak **,** hangilerini **index** olarak görmek istediğimizi yazıyoruz... **:**

```python
tips.pivot_table(['tip_pct', 'size'] , index=['time', 'day'] , columns='smoker')
```

+ Yukarıdaki işlemi **.groupby( )** ile yapsaydık **,** ancak aşağıdaki gibi yapabilirdik **:**

```python
tips.groupby(['time' , 'day' , 'smoker'])[['size' , 'tip_pct']].mean()
```

+ #### Bazen de Tablolarda **,** TÜM (ALL) Veriler için de bir sonuç görmek isteyebiliriz... O zaman da **"``(margins=True)``"** komutunu (Parametresini) kullanmalıyız... **:**

```python
tips.pivot_table(['tip_pct', 'size'] , index=['time', 'day'] , columns='smoker' , margins=True)
```

+ **NaN** değerler yerine de başka bir değer yazmak istersek **,** mesela her birine **'0'** değeri atamak istersek de **``(fill_value=0)``** Parametresini kullanmalıyız... **:**
+ Ayrıca **;** default olarak **``(aggfunc='mean')``** olan Parametreyi değiştirmek için de **,** aşağıdaki gibi hangi matematiksel fonksiyonu kullanmak istiyorsak onu ekleyebiliriz... **:**

```python
tips.pivot_table('tip_pct', index=['time', 'size', 'smoker'] , columns='day' , aggfunc='var' ,
                 fill_value=0 , margins=True)
```
